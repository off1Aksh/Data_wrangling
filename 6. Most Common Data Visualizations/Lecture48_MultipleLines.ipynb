```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
data=pd.read_csv("happy2015.csv")
```

```python
data.head()
```

#### Multiple lines in 1 graph

##### visualize family, freedom & generosity

```python
plt.subplot()
data.Family.plot(kind = 'line', color = 'g',label = 'Family',linewidth=1,linestyle = '-')
data.Freedom.plot(color = 'r',label = 'Freedom',linewidth=1, linestyle = '-')
data.Generosity.plot(color = 'b',label = 'Generosity',linewidth=1,linestyle = '-')
plt.legend(loc='best')
plt.xlabel("Countries")
plt.title('Line Plot')
plt.show()
```

### Export visualization

```python
#Importing the required packages

from bokeh.io import output_file, show

from bokeh.plotting import figure

#Creating our data arrays used for plotting the line plot

x = [5,6,7,8,9,10]

y = [1,2,3,4,5,6]

#Calling the figure() function to create the figure of the plot

plot = figure()

#Creating a line plot using the line() function

plot.line(x,y)

#Creating markers on our line plot at the location of the intersection between x and y

plot.cross(x,y, size = 15)

#Output the plot 

output_file('line_plot.html')

show(plot)
```
