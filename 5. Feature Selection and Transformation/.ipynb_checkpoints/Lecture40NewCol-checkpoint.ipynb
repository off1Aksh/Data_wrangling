```python
import pandas as pd
```

```python
n=pd.read_csv("nobel.csv")
```

```python
n.head()
```

## Age at the Timing of Winning

```python
n["Birth Date"].value_counts().head()
```

```python
n["Birth Date"] = n["Birth Date"].replace(to_replace="nan", value=0)
## remove NAs
```

### Extract birth year

```python
n["Birth Year"] = n["Birth Date"].str[0:4] ##extract the first 4 digits-
## birth year
```

```python
n["Birth Year"].head()
```

```python
n.head()
```

### Convert Birth year to Numeric

```python
n["Birth Year"] = pd.to_numeric(n["Birth Year"])
```

```python
n["Birth Year"].head()
```

#### winning age

```python
n["Age"] = n["Year"] - n["Birth Year"]
```

```python
n.Age.describe() ## stats for age
```
