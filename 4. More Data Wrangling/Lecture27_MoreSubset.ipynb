```python
nfa.UN_subregion.unique() ## list all unique categorical values in a column
```

### subset based on row values

```python
seac=nfa[nfa.UN_subregion == 'South-Eastern Asia']
```

```python
seac.head()
```

### subset on basis of row numerical values

```python
seac_fl=seac[seac.forest_land >= 5] ## isolate forest area>=5
```

```python
seac_fl.head()
```

### subset based on list

```python
years = [1990, 2000, 2005] ## list of years
```

```python

seac.year.isin(years)
```

```python
seac_y = seac[seac.year.isin(years)]
```

```python
seac_y.head()
```

### Select Rows of Pandas Dataframe Based on Values NOT in a list

```python
nfa.UN_region.unique()
```

```python
continents=['Africa', 'Latin America and the Caribbean',
       'Oceania', 'North America', 'World'] ##we want to leave these out
```

```python
nfa_cont = nfa[~nfa.UN_region.isin(continents)] ## check if members of the
##column UN_region are in the list
## we want to only retain data for Asia & africa
```

```python
nfa_cont.head()
```

```python
nfa_cont2 = nfa[nfa.UN_region.isin(continents)]
```

```python
nfa_cont2.head()
```
